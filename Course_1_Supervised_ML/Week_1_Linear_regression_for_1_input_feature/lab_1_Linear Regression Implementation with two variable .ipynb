```python
# as in this we only have two data points 
# we have defined array of two data points where x_train is an input feature and y_train is an output feature 
X_train = np.array([1.0,2.0])
Y_train = np.array([300.0,500.0])
print(f"Input feature for tranning are:  {X_train}")
print(f"output feature for tranning are: {Y_train}")
print("Both are map correctly")
```

# Model Function
ùëìùë§,ùëè(ùë•(ùëñ))=ùë§ùë•(ùëñ)+ùëè(1)

The formula above is how you can represent straight lines - different values of ùë§ and ùëè give you different straight lines on the plot.

Let's try to get a better intuition for this through the code blocks below. Let's start with  ùë§=100 and  ùëè=100

```python
#defining the parameters hear, we are choosing the parameters manually by intusion 
#generally in ml there parametres are automatically made by Alogrithms but for simplification defining it manually 
# i have tried value like 150,100 190,150etc at last it was 200,100
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
# Defining the function to train a model 
def Compute_Yhat(x,w,b): #yhat means predit value of y 
    length_of_input= len(x)
    f_wb= np.zeros(length_of_input)
    for i in range(0,length_of_input):
        f_wb[i] = w*x[i]+b
    print("function returned")
    return f_wb
print("function created sucessfully ")
```

```python
#now lets call compute model function with our patameters to train our linear line 
model_f_wb = Compute_Yhat(X_train,w,b)
print(model_f_wb)
```

```python
#lets plot our trained model and our datapoints

#ploting our model's pridection 
plt.plot(X_train,model_f_wb,c="blue",label="our model")

#ploting our datapoints 
plt.scatter(X_train,Y_train,marker="x",c="red",label="Datapoints")

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend() 
plt.show()
```

```python
# as now we have fited the linear regession completely so its time to predit
#lets calculate the cost of house whose area is 1200 sqft
x=np.array([1.2])
yhat = Compute_Yhat(x,w,b)
print(f"The predit cost of house is {yhat}$")
```
