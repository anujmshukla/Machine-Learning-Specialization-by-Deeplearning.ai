## Lets plot the data in which there is only one input variable or one input feature 

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])

X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

## Lets plot the data in which there is only multiple input variable or multiple input feature 

```python
pos1= y_train == 1
neg1 = y_train == 0

pos2 = y_train2 == 1
neg2 = y_train2 == 0


_,ax = plt.subplots(1,2,figsize=(8,3))

ax[0].scatter(X_train2[pos2,0], X_train2[pos2,1], marker="x", label="y=1", c="red", s=80 )
ax[0].scatter(X_train2[neg2,0], X_train2[neg2,1], marker="o", label="y=0", c="blue", s=100 )
ax[0].set_title("Two Feature")
ax[0].set_xlabel("X 1")
ax[0].set_ylabel("X 2")

ax[1].scatter(x_train[pos1],y_train[pos1],marker='x',c="red",label="y=1",s=80)
ax[1].scatter(x_train[neg1],y_train[neg1],marker='o',c="blue",label="y=0", s=100)
ax[1].set_title("Single Feature")
ax[1].set_xlabel("X_Train")
ax[1].set_ylabel("Y_Train")
plt.show()
```
